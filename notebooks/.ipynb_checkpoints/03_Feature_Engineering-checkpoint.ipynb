```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("/100_Days_of_ml/Hybrid_Movie_Recommendation_System/data/tmdb_5000_movies.csv")
credits = pd.read_csv("/100_Days_of_ml/Hybrid_Movie_Recommendation_System/data/tmdb_5000_credits.csv")
```

```python
movies = movies[
    [
        'id',
        'title',
        'overview',
        'genres',
        'keywords'
    ]
]

credits = credits[
    [
        'movie_id',
        'cast',
        'crew'
    ]
]
```

```python
movies = movies.merge(
    credits,
    left_on="id",
    right_on="movie_id"
)
```

```python
movies.drop(
    "movie_id",
    axis=1,
    inplace=True
)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
import ast
```

```python
movies["genres"] = movies["genres"].apply(ast.literal_eval)
```

```python
type(movies.iloc[0].genres)
```

```python
movies.iloc[0].genres
```

```python
movies["keywords"] = movies["keywords"].apply(ast.literal_eval)

movies["cast"] = movies["cast"].apply(ast.literal_eval)

movies["crew"] = movies["crew"].apply(ast.literal_eval)
```

```python
print(type(movies.iloc[0].keywords))
print(type(movies.iloc[0].cast))
print(type(movies.iloc[0].crew))
```

```python
## this is the function used to convert we have to extract only names in movies["genres"]

def converrt(obj):
    L = []

    for i in obj:
        L.append(i["name"])

    return L
```

```python
movies["genres"] = movies["genres"].apply(converrt)
```

```python
movies.iloc[0]["genres"]
```

```python
movies["keywords"] = movies["keywords"].apply(converrt)
```

### Function For Extraction only 3 top(lead) Actors Of Any Movie

```python
def Fetch_Cast(obj):
    L = []
    counter = 0
    for i in obj:
        if counter != 3:
            L.append(i["name"])
            counter += 1

        else:
            break
    return L
```

```python
movies["cast"] = movies["cast"].apply(Fetch_Cast)
```

```python
movies.iloc[2]["cast"]
```

### Extract Director From Crew

```python
def fetch_director(obj):

    L = []

    for i in obj:

        if i["job"] == "Director":
            L.append(i["name"])
            break

    return L
```

```python
movies["crew"] = movies["crew"].apply(fetch_director)
```

```python
movies.iloc[1]["crew"]
```

## Remove Spaces

```python
def RemoveSpaces(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies["genres"] = movies["genres"].apply(RemoveSpaces)
movies["keywords"] = movies["keywords"].apply(RemoveSpaces)
movies["cast"] = movies["cast"].apply(RemoveSpaces)
movies["crew"] = movies["crew"].apply(RemoveSpaces)
```

```python
movies.iloc[0]
```

### Creating Tags

```python
movies["overview"] = movies["overview"].apply(lambda x:x.split())
```

```python
movies.iloc[0]["overview"]
type(movies.iloc[0]["overview"])
```

```python
movies["tags"] = movies["overview"] + \
                 movies["genres"] + \
                 movies["keywords"] + \
                 movies["cast"] + \
                 movies["crew"]
```

```python
movies.iloc[20]["tags"]
```

```python
# We keep Only These Columns
    # id → useful for fetching movie details or posters.
    # title → user searches by movie title.
    # tags → used to calculate movie similarity.

movies = movies[
    [
        "id",
        "title",
        "tags"
    ]
]
```

```python
movies.head()
```

```python
movies["tags"] = movies["tags"].apply(lambda x:" ".join(x))
```

```python
movies.iloc[0]["tags"]
```

```python
movies["tags"] = movies["tags"].apply(lambda x: x.lower())
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
print(ps.stem("loving"))
print(ps.stem("travelling"))
```

### Writing Function For Stemming

```python
# action adventure sciencefiction ->>> before split
# Action
# Adventure ->> After Split
# sciencefiction  
 
def stem(text):
    y = [] 
    for i in text.split(): # split texts 
        y.append(ps.stem(i)) # Check For all Text One by one  
    return " ".join(y) # again convert to sentence 
```

```python
movies["tags"]  = movies["tags"].apply(stem)
```

```python
movies.iloc[0]["tags"]
```

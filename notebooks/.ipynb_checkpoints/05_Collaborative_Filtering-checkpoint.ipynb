```python
movie_ratings = ratings.merge(
                movies,
                on="movieId"
)
```

```python
movie_ratings.head()
```

### Make Pivot table

```python
movie_matrix = movie_ratings.pivot_table(
                index = "userId",
                columns = "title",
                values = "rating"
)
```

```python
movie_matrix.tail(5)
```

```python
movie_matrix.shape
```

```python
movie_matrix.isnull().sum()
```

```python
print(movie_matrix.isnull().sum().sum())
```

### All Nan filled with zeroes 

```python
movie_matrix_filled = movie_matrix.fillna(0)
```

```python
movie_matrix_filled.head()
```

```python
print(movie_matrix_filled.isnull().sum().sum())
```

### We Can Compare Movies

```python
# for this we want rows-> movies and columns -> users for this we use transpose
movie_matrix_T = movie_matrix_filled.T
```

```python
movie_matrix_T.head()
```

```python
# Before Transpose
movie_matrix_filled.shape
```

```python
# After Transpose
movie_matrix_T.shape
```

# Now For Comparing Movies We use Cosine Similiarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(movie_matrix_T)
```

```python
similarity # This Gives Numpy Array and we have to convert it in dataframe 
```

```python
similarity.shape
```

```python
similarity_df = pd.DataFrame(
    similarity,
    index=movie_matrix_T.index,
    columns=movie_matrix_T.index
)
```

```python
similarity_df.head()
```

```python
similarity_df["Toy Story (1995)"].sort_values(
    ascending=False
).head(10)
```

```python
def recommend_movie(movie_name):

    if movie_name not in similarity_df.columns:
        return "Movie Not Found"

    similar_movies = similarity_df[movie_name]

    similar_movies = similar_movies.sort_values(
        ascending=False
    )

    return similar_movies.iloc[1:6]
```

```python
recommend_movie("Zulu (2013)")
```

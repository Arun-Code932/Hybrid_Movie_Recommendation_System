```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("/100_Days_of_ml/Hybrid_Movie_Recommendation_System/data/tmdb_5000_movies.csv")
credits = pd.read_csv("/100_Days_of_ml/Hybrid_Movie_Recommendation_System/data/tmdb_5000_credits.csv")
```

```python
movies = movies[
    [
        'id',
        'title',
        'overview',
        'genres',
        'keywords'
    ]
]

credits = credits[
    [
        'movie_id',
        'cast',
        'crew'
    ]
]
```

## Merge DataFrames

```python
movies = movies.merge(
    credits,
    left_on="id",
    right_on="movie_id"
)
```

```python
movies.shape
```

```python
movies.head()
```

```python
# we have both id and movie id is same then we have to remove one because both are not necessary
movies.drop(
    "movie_id",
    axis=1,
    inplace=True
)
```

```python
movies.head()
```

```python
movies.columns
```

```python
movies.isnull().sum()
```

### Remove Missing Rows

```python
# Why Are We Removing Rows Instead of Filling Them?
# Only a few rows have missing overview values.
# Since the dataset has about 4800 movies, removing 3 rows has almost no impact.
# If thousands of rows were missing, we'd consider filling the missing values instead.b
```

```python
movies.dropna(inplace = True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum() # if found duplicates then remove by ->>  movies.drop_duplicates(inplace=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
vectors = cv.fit_transform(movies["tags"]).toarray()
```

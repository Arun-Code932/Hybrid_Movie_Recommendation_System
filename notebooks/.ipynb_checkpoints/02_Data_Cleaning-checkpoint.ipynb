```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("/100_Days_of_ml/Hybrid_Movie_Recommendation_System/data/tmdb_5000_movies.csv")
credits = pd.read_csv("/100_Days_of_ml/Hybrid_Movie_Recommendation_System/data/tmdb_5000_credits.csv")
```

```python
movies = movies[
    [
        'id',
        'title',
        'overview',
        'genres',
        'keywords'
    ]
]

credits = credits[
    [
        'movie_id',
        'cast',
        'crew'
    ]
]
```

## Merge DataFrames

```python
movies = movies.merge(
    credits,
    left_on="id",
    right_on="movie_id"
)
```

```python
movies.shape
```

```python
movies.head()
```

```python
# we have both id and movie id is same then we have to remove one because both are not necessary
movies.drop(
    "movie_id",
    axis=1,
    inplace=True
)
```

```python
movies.isnull().sum()
```

### Remove Missing Rows

```python
# Why Are We Removing Rows Instead of Filling Them?
# Only a few rows have missing overview values.
# Since the dataset has about 4800 movies, removing 3 rows has almost no impact.
# If thousands of rows were missing, we'd consider filling the missing values instead.b
```

```python
movies.dropna(inplace = True)
```

### It simply converted text into a Python list -> ast.literal_eval()

```python
import ast
```

```python
movies["genres"] = movies["genres"].apply(ast.literal_eval)
```

```python
type(movies.iloc[0].genres)
```

```python
movies.iloc[0].genres
```

```python
movies["keywords"] = movies["keywords"].apply(ast.literal_eval)

movies["cast"] = movies["cast"].apply(ast.literal_eval)

movies["crew"] = movies["crew"].apply(ast.literal_eval)
```

```python
print(type(movies.iloc[0].keywords))
print(type(movies.iloc[0].cast))
print(type(movies.iloc[0].crew))
```

### Feature Extraction

```python
## this is the function used to convert we have to extract only names in movies["genres"]

def converrt(obj):
    L = []

    for i in obj:
        L.append(i["name"])

    return L
```

```python
movies["genres"] = movies["genres"].apply(converrt)
```

```python
movies.iloc[0]["genres"]
```

```python
movies["keywords"] = movies["keywords"].apply(converrt)
```

### Function For Extraction only 3 top(lead) Actors Of Any Movie

```python
def Fetch_Cast(obj):
    L = []
    counter = 0
    for i in obj:
        if counter != 3:
            L.append(i["name"])
            counter += 1

        else:
            break
    return L
```

```python
movies["cast"] = movies["cast"].apply(Fetch_Cast)
```

```python
movies.iloc[2]["cast"]
```

### Extract Director From Crew

```python
def fetch_director(obj):

    L = []

    for i in obj:

        if i["job"] == "Director":
            L.append(i["name"])
            break

    return L
```

```python
movies["crew"] = movies["crew"].apply(fetch_director)
```

```python
movies.iloc[1]["crew"]
```
